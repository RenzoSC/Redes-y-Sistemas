# Limpieza de datos

## Las siguientes lineas las utilizamos para limpiar datos que nos sean de interes.

```python
time_gen = data['vectime'].loc[[23]].tolist()
buffer_gen = data['vecvalue'].loc[[23]].tolist()

time_queue = data['vectime'].loc[[24]].tolist()
buffer_queue = data['vecvalue'].loc[[24]].tolist()

time_sink = data['vectime'].loc[[25]].tolist()
buffer_sink = data['vecvalue'].loc[[25]].tolist()

print(time_gen)
print(buffer_gen)
print(time_queue)
print(buffer_queue)
print(time_sink)
print(buffer_sink)

```

```python
# Convertir las cadenas de tiempo en listas de valores flotantes
# Dividir los elementos de la lista con comas
gen_elems_by_coma = ','.join(time_gen)

# Convertir la cadena de valores separados por comas en una lista
gen_elems_values = gen_elems_by_coma.split(',')

# Imprimir la lista resultante
print(gen_elems_values)
time_queue = list(map(float, time_queue))
time_sink = list(map(float, time_sink))

# Convertir las cadenas de buffer en listas de enteros
buffer_gen = list(map(int, buffer_gen))
buffer_queue = list(map(int, buffer_queue))
buffer_sink = list(map(int, buffer_sink))

# Imprimir las listas resultantes
print(time_gen)
print(buffer_gen)
print(time_queue)
print(buffer_queue)
print(time_sink)
print(buffer_sink)

```
